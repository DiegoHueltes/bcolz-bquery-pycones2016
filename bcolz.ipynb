# Leer, procesar y almacenar (bcolz y hdf5)

```python
import pandas as pd
import bcolz
from bquery import ctable
```

```python
# Puedes descargarlo de https://nyctaxitrips.blob.core.windows.net/data/trip_data_1.csv.zip
df = pd.read_csv('trip_fare_1.csv', names=['medallion', 'hack_license', 'vendor_id', 'pickup_datetime', 
                                           'payment_type','fare_amount', 'surcharge', 'mta_tax', 'tip_amount', 
                                           'tolls_amount', 'total_amount'], header=0)
```

```python
df.to_hdf('trip_fare_1.h5', 'df')
```

```python
ctable.fromdataframe(df, rootdir='trip_fare_1.bcolz', mode='w')
```

```python
del df
```

# Query

```python
def pandas_test(df):
    return df.eval('cos((3+tip_amount)/sqrt(2*fare_amount))')
    
def pandas_test_2():
    df2 = pd.read_hdf('trip_fare_1.h5', 'df')
    results = df.eval('cos((3+tip_amount)/sqrt(2*fare_amount))')
    del df
    return results
    
def bcolz_test():
    ct = ctable(rootdir='trip_fare_1.bcolz', mode='r')
    return ct.eval('cos((3+tip_amount)/sqrt(2*fare_amount))', out_flavor="numpy")
    
```
